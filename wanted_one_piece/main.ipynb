```python
# packages
import cv2
import matplotlib.pyplot as plt
# import matplotlib.image as mpimg
from skimage import io
import numpy as np
```

```python
# upload image
# file = mpimg.imread('images/image.jpg')
img = cv2.imread("images/image.jpg")
io.imshow(img)
plt.show()
```

```python
# you must find the edge in an image
# let's use the function cv2.adaptiveThreshold()
df = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
df_blur = cv2.medianBlur(df, 3)
io.imshow(img)
plt.show()
```

```python
edges = cv2.adaptiveThreshold(df_blur, 250, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 75, 3)
io.imshow(edges)
plt.show()
```

```python
# d = diameter of each pixel
# sigma color = big number, larger areas equals
# sigma space = dad point can influency others if the number is larger

blurred_img = cv2.bilateralFilter(img, d=21, sigmaColor=250,sigmaSpace=250)
io.imshow(blurred_img)
plt.show()
```

```python
cartoon = cv2.bitwise_and(blurred_img, blurred_img, mask=edges)
io.imshow(cartoon)
plt.show()
io.imshow(cartoon)
plt.savefig('cartoon/result.png')
```
